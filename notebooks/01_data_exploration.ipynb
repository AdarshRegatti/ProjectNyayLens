```python
"""
Phase 1: Dataset Overview Analysis
Purpose: Understand the distribution and characteristics of our legal judgment dataset
Author: [Your Name]
Date: December 7, 2025
"""

import sys
from pathlib import Path
from collections import defaultdict

BASE_DIR = Path("../data/raw")
def count_pdfs_by_year():
    """Count PDFs in each year folder (top-level folders under BASE_DIR)."""
    base = Path(BASE_DIR)
    if not base.exists():
        raise FileNotFoundError(f"BASE_DIR does not exist: {base.resolve()}")
    if not base.is_dir():
        raise NotADirectoryError(f"BASE_DIR is not a directory: {base.resolve()}")

    year_counts = {}

    # Get all year folders (only directories directly under BASE_DIR)
    year_folders = [f for f in base.iterdir() if f.is_dir()]

    # Sort: prefer numeric year if possible, otherwise lexicographic
    def sort_key(p):
        try:
            return (0, int(p.name))   # numeric years first (as integers)
        except Exception:
            return (1, p.name)        # then non-numeric names

    year_folders = sorted(year_folders, key=sort_key)

    print("="*60)
    print("COUNTING PDFs BY YEAR")
    print("="*60)

    if not year_folders:
        print("No subfolders found under:", base)
        return year_counts

    for year_folder in year_folders:
        year = year_folder.name

        # Count PDF files directly inside this folder (not recursive)
        pdf_files = list(year_folder.glob("*.pdf"))
        count = len(pdf_files)

        year_counts[year] = count
        print(f"Year {year}: {count} PDFs")

    return year_counts

def analyze_distribution(year_counts):
    """Analyze the distribution patterns."""
    print("\n" + "="*60)
    print("DISTRIBUTION ANALYSIS")
    print("="*60)

    if not year_counts:
        print("No data to analyze.")
        return {
            'total': 0,
            'average': 0,
            'decades': {},
            'anomalies': {}
        }

    # Basic statistics
    total = sum(year_counts.values())
    avg = total / len(year_counts) if len(year_counts) > 0 else 0

    print(f"\nTotal PDFs: {total:,}")
    print(f"Total Years: {len(year_counts)}")
    print(f"Average per year: {avg:.1f}")

    # Find peaks and valleys
    max_year = max(year_counts, key=year_counts.get)
    min_year = min(year_counts, key=year_counts.get)

    print(f"\nHighest: {max_year} with {year_counts[max_year]} PDFs")
    print(f"Lowest: {min_year} with {year_counts[min_year]} PDFs")

    # Decade-wise summary (only for numeric year folders)
    print("\n" + "="*60)
    print("DECADE-WISE SUMMARY")
    print("="*60)

    decades = defaultdict(int)
    for year, count in year_counts.items():
        try:
            decade = (int(year) // 10) * 10
            decades[decade] += count
        except Exception:
            # skip non-numeric folder names for decade summary
            continue

    if decades:
        for decade in sorted(decades.keys()):
            print(f"{decade}s: {decades[decade]:,} PDFs")
    else:
        print("No numeric year folders found for decade summary.")

    # Identify anomalies (less than 100 PDFs)
    print("\n" + "="*60)
    print("ANOMALIES (< 100 PDFs)")
    print("="*60)

    anomalies = {year: count for year, count in year_counts.items() if count < 100}
    if anomalies:
        for year, count in sorted(anomalies.items()):
            print(f"{year}: {count} PDFs ⚠️")
    else:
        print("No anomalies found (all years >= 100 PDFs).")

    return {
        'total': total,
        'average': avg,
        'decades': dict(decades),
        'anomalies': anomalies
    }

def estimate_processing_time(total_pdfs):
    """Estimate how long processing will take."""
    print("\n" + "="*60)
    print("PROCESSING TIME ESTIMATES")
    print("="*60)

    scenarios = [
        ("Fast (1 sec/PDF)", 1),
        ("Moderate (3 sec/PDF)", 3),
        ("Slow (5 sec/PDF)", 5),
        ("Complex (10 sec/PDF)", 10)
    ]

    for scenario_name, seconds_per_pdf in scenarios:
        total_seconds = total_pdfs * seconds_per_pdf
        hours = total_seconds / 3600
        print(f"{scenario_name}: {hours:.1f} hours ({total_seconds/60:.0f} minutes)")

def main():
    """Main execution function."""
    print("\n" + "="*70)
    print(" "*15 + "LEGAL JUDGMENT DATASET ANALYSIS")
    print(" "*20 + "Phase 1: Overview")
    print("="*70 + "\n")

    # Step 1: Count PDFs
    year_counts = count_pdfs_by_year()

    # Step 2: Analyze distribution
    stats = analyze_distribution(year_counts)

    # Step 3: Processing estimates
    estimate_processing_time(stats['total'])

    print("\n" + "="*70)
    print("✓ Analysis complete!")
    print("="*70 + "\n")

    return year_counts, stats

# Run the analysis
if __name__ == "__main__":
    try:
        year_counts, stats = main()
    except Exception as e:
        print("ERROR:", e, file=sys.stderr)
        raise
```

```python
"""
Phase 1: PDF Quality Check
Purpose: Sample random PDFs to check text extractability
Author: [Your Name]
Date: December 7, 2025
"""

import random
from pathlib import Path
import PyPDF2
import sys

# Configuration - Get absolute path
SCRIPT_DIR = Path.cwd()
PROJECT_ROOT = SCRIPT_DIR.parent
BASE_DIR = PROJECT_ROOT / "data" / "raw"

SAMPLE_SIZE = 10  # Check 10 random PDFs

def get_all_pdfs():
    """Get list of all PDF files"""
    print(f"Looking for PDFs in: {BASE_DIR}")
    
    if not BASE_DIR.exists():
        print(f"❌ ERROR: Directory does not exist: {BASE_DIR}")
        sys.exit(1)
    
    all_pdfs = list(BASE_DIR.rglob("*.pdf"))
    return all_pdfs

def check_pdf_quality(pdf_path):
    """Check if PDF has extractable text"""
    
    try:
        with open(pdf_path, 'rb') as file:
            pdf_reader = PyPDF2.PdfReader(file)
            
            num_pages = len(pdf_reader.pages)
            
            # Extract text from first page
            first_page = pdf_reader.pages[0]
            text = first_page.extract_text()
            
            # Check quality
            text_length = len(text.strip())
            has_text = text_length > 100  # At least 100 characters
            
            return {
                'filename': pdf_path.name,
                'year': pdf_path.parent.name,
                'num_pages': num_pages,
                'first_page_chars': text_length,
                'has_extractable_text': has_text,
                'sample_text': text[:200] if has_text else "NO TEXT FOUND"
            }
    
    except Exception as e:
        return {
            'filename': pdf_path.name,
            'year': pdf_path.parent.name,
            'error': str(e)
        }

def main():
    """Main execution"""
    
    print("="*70)
    print(" "*20 + "PDF QUALITY CHECK")
    print("="*70 + "\n")
    
    # Get all PDFs
    all_pdfs = get_all_pdfs()
    print(f"Total PDFs found: {len(all_pdfs):,}\n")
    
    if len(all_pdfs) == 0:
        print("❌ No PDFs found! Check your data/raw directory.")
        return
    
    # Sample random PDFs
    sample_pdfs = random.sample(all_pdfs, min(SAMPLE_SIZE, len(all_pdfs)))
    
    print(f"Checking {len(sample_pdfs)} random PDFs...")
    print("="*70)
    
    results = []
    for i, pdf_path in enumerate(sample_pdfs, 1):
        print(f"\n[{i}/{len(sample_pdfs)}] {pdf_path.parent.name}/{pdf_path.name}")
        print("-"*70)
        
        result = check_pdf_quality(pdf_path)
        results.append(result)
        
        if 'error' in result:
            print(f"❌ ERROR: {result['error']}")
        else:
            print(f"Year: {result['year']}")
            print(f"Pages: {result['num_pages']}")
            print(f"First page characters: {result['first_page_chars']}")
            print(f"Has text: {'✓ YES' if result['has_extractable_text'] else '✗ NO (needs OCR)'}")
            
            if result['has_extractable_text']:
                print(f"\nSample text:")
                print(f"{result['sample_text']}...")
    
    # Summary
    print("\n" + "="*70)
    print("QUALITY SUMMARY")
    print("="*70)
    
    successful = [r for r in results if 'error' not in r]
    with_text = [r for r in successful if r['has_extractable_text']]
    without_text = [r for r in successful if not r['has_extractable_text']]
    errors = [r for r in results if 'error' in r]
    
    print(f"\nTotal checked: {len(results)}")
    print(f"✓ With extractable text: {len(with_text)}")
    print(f"✗ Need OCR: {len(without_text)}")
    print(f"❌ Errors: {len(errors)}")
    
    if len(successful) > 0 and len(with_text) > 0:
        success_rate = (len(with_text) / len(successful)) * 100
        print(f"\n📊 Success rate: {success_rate:.1f}%")
    
    print("\n" + "="*70)
    print("✓ Quality check complete!")
    print("="*70)

if __name__ == "__main__":
    main()
```
